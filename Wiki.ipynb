Extracts the table from the wikipedia article.

```python
import pandas as pd
import numpy as np
from pandas.io.html import read_html
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

wikitable = read_html(page, attrs={"class":"wikitable sortable"})[0]
```

Creates the wikitable as as dataframe.

```python
table = pd.DataFrame(wikitable)
table
table.head()
```

Takes only postcodes that have a Borough assigned.

```python
table2 = table[~table.Borough.isin(['Not assigned'])]
```

Groups all of the neighbourhoods by their Postcode and Borough and sets the one borough with a missing neighbourhood to it's borough name.

```python
table3 = pd.DataFrame(table2.groupby(['Postcode','Borough'])['Neighbourhood'].apply(", ".join))
table3.loc[table3['Neighbourhood'].isin(['Not assigned'])] = "Queen's Park"
```
